```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

coronavirus = pd.read_csv(r'C:\Users\Lenovo\Desktop\Coronavirus\Covid19_data.txt')


def getGrup(country1, country2 = None, country3 = None, country4 = None, country5 = None):

    corona = coronavirus.groupby('Country')
    country1_grup = corona.get_group(country1)

    if country2 == None:
        country2_grup = pd.DataFrame({'Date': (country1_grup['Date']), 'Confirmed': list(range(0,len(country1_grup['Date']))) })
    else:
        country2_grup = corona.get_group(country2)

    if country3 == None:
        country3_grup = pd.DataFrame({'Date': (country1_grup['Date']), 'Confirmed': list(range(0,len(country1_grup['Date']))) })
    else:
        country3_grup = corona.get_group(country3)


    if country4 == None:
        country4_grup = pd.DataFrame({'Date': (country1_grup['Date']), 'Confirmed': list(range(0,len(country1_grup['Date']))) })
    else:
        country4_grup = corona.get_group(country4)

    if country5 == None:
        country5_grup = pd.DataFrame({'Date': (country1_grup['Date']), 'Confirmed': list(range(0,len(country1_grup['Date']))) })
    else:
        country5_grup = corona.get_group(country5)


    country1_grup = country1_grup[['Date', 'Confirmed']]
    country2_grup = country2_grup[['Date', 'Confirmed']]
    country3_grup = country3_grup[['Date', 'Confirmed']]
    country4_grup = country4_grup[['Date', 'Confirmed']]
    country5_grup = country5_grup[['Date', 'Confirmed']]


    date_array = np.array(country1_grup['Date'])

    country1_array = np.array(country1_grup['Confirmed'])
    country2_array = np.array(country2_grup['Confirmed'])
    country3_array = np.array(country3_grup['Confirmed'])
    country4_array = np.array(country4_grup['Confirmed'])
    country5_array = np.array(country5_grup['Confirmed'])

    plt.xticks(rotation=90)

    plt.plot(date_array,country1_array, label= country1)
    plt.plot(date_array, country2_array, label= country2)
    plt.plot(date_array, country3_array, label= country3)
    plt.plot(date_array, country4_array, label= country4)
    plt.plot(date_array, country5_array, label= country5)

    plt.xlabel('Dates')
    plt.ylabel('Population')
    plt.legend()
    plt.show()
    #plt.savefig('Plot.png')


indiaGrp = getGrup('China', 'US', 'Spain', 'Pakistan', 'Italy')

print(indiaGrp)







```

```python

class Country():
    
    def __init__(self, filename):
        import pandas as pd
        self.coronavirus = pd.read_csv(filename)
        
        
    def getGrup(self, country1):
        self.country1 = country1
        corona = self.coronavirus.groupby('Country')
        country_grup = corona.get_group(self.country1)
        country_grup = country_grup[['Date', 'Confirmed']]
        
        return country_grup
    
    def plotting(self, country1):
        self.country1 = country1
        corona = self.coronavirus.groupby('Country')
        country_grup = corona.get_group(self.country1)
        country_grup = country_grup[['Date', 'Confirmed']]
        %matplotlib inline
        import matplotlib.pyplot as plt
        plt.plot(country_grup['Date'], country_grup['Confirmed'])
        plt.show()

c = Country(r"C:\Users\Lenovo\Desktop\Coronavirus\Covid19_data.txt")
c.plotting('US')

        
```
